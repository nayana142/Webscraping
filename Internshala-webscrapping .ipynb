## One page

```python
jobs=soup.findAll('div',{'class':'container-fluid individual_internship visibilityTrackerItem'})
job_details=[]
for job in jobs:
    job_detail={}
    x=job.find('div',{'class':'company'})
    company_name=x.find('h4').text.strip()
    job_name=x.find('h3').text.strip()
    job_detail['company_name']=company_name
    job_detail['job_name']=job_name
    
    locations=job.findAll('a',{'class':'location_link view_detail_button'})
    for location in locations:
        job_detail.setdefault('location', [])
        job_detail['location'].append(location.text)

    joining_date=job.find('div',{'class':'item_body'}).text.strip()
    job_detail['joining_date']=joining_date

    salary=job.find('div',{'class':'item_body salary'}).text.strip()
    job_detail['salary']=salary


    experience=job.find('div',{'class':'item_body mobile-text'}).text.strip()
    job_detail['experiance']=experience


    job_post_date=job.find('div',{'class':'status status-small status-inactive'}).text.strip()
    job_detail['job_post_date']=job_post_date
    
#--------------------------------------------------------------------------------------------------
    
    job_link=job.find('div',{'class':'cta_container'}).find('a')['href']
    url1='https://internshala.com'+job_link
    response = requests.get(url1)
    content = response.text
    soup = BeautifulSoup(content,'lxml') 
    
    job_last_date=soup.find('div',{'class':'internship_other_details_container'})
    job_last_dates=job_last_date.findAll('div',{'class':'item_body'})

    for job_last_date in job_last_dates:
        job_last_date=job_last_date.text.strip()
    job_detail['job_last_date']=job_last_date
    
    skills=soup.find('div',{'class':'round_tabs_container'})
    skills=skills.findAll('span')
    for skill in skills:
        job_detail.setdefault('skills', [])
        job_detail['skills'].append(skill.text)
        
    no_of_openings=soup.findAll('div',{'class':'text-container'})
    for no_of_opening in no_of_openings:
        no_of_opening=no_of_opening.text.strip()
    job_detail['no_of_opening']=no_of_opening
    
    job_details.append(job_detail)

import pandas as pd
df=pd.DataFrame(job_details)
df
```

# All page

```python
from tqdm import tqdm
job_details=[]
for i in tqdm(range(1,146)):
    url=f'https://internshala.com/jobs/page-{i}/'
    response = requests.get(url)
    content = response.text
    soup = BeautifulSoup(content,'lxml') 
    jobs=soup.findAll('div',{'class':'container-fluid individual_internship visibilityTrackerItem'})

    for job in jobs:
        job_detail={}
        x=job.find('div',{'class':'company'})
        company_name=x.find('h4').text.strip()
        job_name=x.find('h3').text.strip()
        job_detail['company_name']=company_name
        job_detail['job_name']=job_name
        locations=job.findAll('a',{'class':'location_link view_detail_button'})
        for location in locations:
            job_detail.setdefault('location', [])
            job_detail['location'].append(location.text)

        joining_date=job.find('div',{'class':'item_body'}).text.strip()
        job_detail['joining_date']=joining_date

        salary=job.find('div',{'class':'item_body salary'}).text.strip()
        job_detail['salary']=salary


        experience=job.find('div',{'class':'item_body mobile-text'}).text.strip()
        job_detail['experiance']=experience


        job_post_date=job.find('div',{'class':'status status-small status-inactive'}).text.strip()
        job_detail['job_post_date']=job_post_date
        
#--------------------------------------------------------------------------------------------------

        job_link=job.find('div',{'class':'cta_container'}).find('a')['href']
        url1='https://internshala.com'+job_link
        response = requests.get(url1)
        content = response.text
        soup = BeautifulSoup(content,'lxml') 

        job_last_date=soup.find('div',{'class':'internship_other_details_container'})
        job_last_dates=job_last_date.findAll('div',{'class':'item_body'})

        job_last_date=soup.find('div',{'class':'internship_other_details_container'})
        if job_last_date is not None:

            job_last_dates=job_last_date.findAll('div',{'class':'item_body'})

            for job_last_date in job_last_dates:
                job_last_date=job_last_date.text.strip()
            job_detail['job_last_date']=job_last_date
        else:
            job_detail['job_last_date']=None


        skills=soup.find('div',{'class':'round_tabs_container'})
        if job_last_date is not None:

            skills=skills.findAll('span')
            for skill in skills:
                job_detail.setdefault('skills', [])
                job_detail['skills'].append(skill.text)
        else:
            job_detail['skills']=None

        no_of_openings=soup.findAll('div',{'class':'text-container'})
        for no_of_opening in no_of_openings:
            no_of_opening1=no_of_opening.text.strip()
        job_detail['no_of_opening']=no_of_opening1

        job_details.append(job_detail)
        
import pandas as pd
df=pd.DataFrame(job_details)
df
```
