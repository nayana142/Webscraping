## Scraping data from one page

```python
clg_details=[]
from tqdm import tqdm
for i in tqdm(range(1,1000)):
    url=f'https://www.indcareer.com/find/all-colleges?page={i}/'
    response = requests.get(url)
    content = response.text
    soup = BeautifulSoup(content,'lxml')
    title=soup.findAll(attrs={'class':"list-group"})
    for i in title:
        clg_info={}
        clg_title=soup.find('h4').text.replace('\n',"").replace(' ',"")
        clg_info['Clg_Name']=clg_title
        affiliated=soup.find('small').text.replace('\n',"").replace(' ',"")
        clg_info['Affiliated']=affiliated
        location=soup.find(attrs={'class':"text-success"}).text.replace('\t',"").replace(' ' ,"").replace('\n',"")
        clg_info['location']=location
        course=soup.find(attrs={'class':"list-unstyled"}).text.replace('\n',"").replace('\t',"").replace(' ' ," ")
        clg_info['course']=course
        clg_details.append(clg_info)
import pandas as pd
df=pd.DataFrame(clg_details)
df.to_csv('indcareer.csv')
```
